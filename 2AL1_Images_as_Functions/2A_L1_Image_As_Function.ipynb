```python
# For given matrix quantize all values of matrix to given min and max integer values

import math

def Quantize_matrix(mat, min_val = 0, max_val = 255):
  new_mat = mat[:]
  for r in range(len(mat)):
    for c in range(len(mat[0])):
      if(math.floor(mat[r][c]) < min_val):
        new_mat[r][c] = min_val
      elif(math.floor(mat[r][c]) > max_val):
        new_mat[r][c] = max_val
      else:
        new_mat[r][c] = math.floor(mat[r][c])
  return new_mat



mat = [[2.5,0.7,3,6], [3.7, 4.5, 1.9, 3.2], [-1.3, 5.2,7.5, 2.9]]
new_mat = Quantize_matrix(mat,0,5)
print(new_mat)
```

```python
# scalar multiplication 
# please upload dolphine and bicycle image first
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

def scale(img, scale):
  return img * scale

img1_path = "/content/Cycle.png"
img2_path = "/content/Dolphin.png"

img1 = cv2.imread(img1_path)
img1_gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

img2 = cv2.imread(img2_path)
img2_gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)

img1_less1 =  scale(img1_gray , 0.5)
img1_More1 =  scale(img1_gray , 1.5)

img1_hori = np.concatenate((img1_gray,img1_less1,img1_More1), axis=1)
cv2_imshow(img1_hori)

img2_less1 =  scale(img2_gray , 0.5)
img2_More1 =  scale(img2_gray , 1.5)

img2_hori = np.concatenate((img2_gray,img2_less1,img2_More1), axis=1)
cv2_imshow(img2_hori)
```
