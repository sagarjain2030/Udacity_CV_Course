```python
# Convert chilli and pepper image to edge image+
# upload image chilli_Pepper.png
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img_path = "/content/Chilli_Pepper.png"
col_img = cv2.imread(img_path)
cv2_imshow(col_img) 
bnw_img = cv2.cvtColor(col_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(bnw_img) 
print(bnw_img.shape)

# cv2 OTSU automatically searches for best possible threshold for CANNY.
threshold_value, _ = cv2.threshold(bnw_img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Using Canny filter to extract edges
edges = cv2.Canny(bnw_img, 0.75 * threshold_value, threshold_value)
cv2_imshow(edges) 
```

```python
# upload image chilli_Pepper.png
# plotting intensity as height for an image
# use 3D surface plot to achieve this
import cv2
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Load the image
img_path = "/content/Chilli_Pepper.png"
col_img = cv2.imread(img_path)
cv2_imshow(col_img) 

# Convert the image to grayscale
gray = cv2.cvtColor(col_img, cv2.COLOR_BGR2GRAY)

# Create x and y arrays with the same shape as the image
y, x  = np.mgrid[0:gray.shape[0], 0:gray.shape[1]]
# Create a figure and a 3D axes object
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the intensity of the image as a height map
ax.plot_surface(x, y, gray, cmap='gray')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Intensity')
plt.show()
```

```python
## Calculate finite difference for zebra
## upload zebra.jpg before running this code

import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Load the image
# Load the image
img_path = "/content/zebra_face.jpg"
col_img = cv2.imread(img_path)
cv2_imshow(col_img) 
gray = cv2.cvtColor(col_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

# Calculate the horizontal and vertical finite difference using the Sobel operator
dx = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize = 3)
dy = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize = 3)

# Calculate the magnitude and direction of the gradient
magnitude = np.sqrt(dx**2 + dy**2)
direction = np.arctan2(dy, dx)

# Display the results
cv2_imshow(magnitude) 
```

```python
## Calculate finite difference for tiger in separate direction
## upload tiger.jpg before running this code
# Remember Sobel operator by default performs correlation

import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img_path = "/content/tiger_face.jpg"
img = cv2.imread(img_path)
cv2_imshow(img)

## converting image to gray scale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

## Sobel operatox X direction
sobel_x_3 = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=3)
sobel_x_9 = cv2.Sobel(gray,cv2.CV_64F,1,0,ksize=9)
## Sobel operatox Y direction
sobel_y_3 = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=3)
sobel_y_9 = cv2.Sobel(gray,cv2.CV_64F,0,1,ksize=9)

Hori_3 = np.concatenate((sobel_x_3, sobel_y_3), axis=1)
Hori_9 = np.concatenate((sobel_x_9, sobel_y_9), axis=1)
cv2_imshow(Hori_3)
cv2_imshow(Hori_9)


# Calculate the magnitude of the gradient
magnitude_3 = np.sqrt(sobel_x_3**2 + sobel_y_3**2)
magnitude_9 = np.sqrt(sobel_x_9**2 + sobel_y_9**2)
Magnitude = np.concatenate((magnitude_3, magnitude_9), axis=1)
cv2_imshow(Magnitude)
```
