```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# please upload panda image
img_path = "/content/panda.jpg"
img = cv2.imread(img_path)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(img_gray.shape)

kernel1 = cv2.getGaussianKernel(11,1)
kernel2 = cv2.getGaussianKernel(11,3)
kernel3 = cv2.getGaussianKernel(11,10)

img_gray1 = cv2.filter2D(img_gray, -1, kernel1)
img_gray2 = cv2.filter2D(img_gray, -1, kernel2)
img_gray3 = cv2.filter2D(img_gray, -1, kernel3)


Hori = np.concatenate((img_gray, img_gray1, img_gray2, img_gray3), axis=1)
cv2_imshow(Hori)
```
