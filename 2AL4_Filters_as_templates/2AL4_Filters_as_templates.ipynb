```python
# Please upload pepper.jpeg image and filter.jpeg before executing code
import numpy as np
import cv2
from google.colab.patches import cv2_imshow

img_path = "/content/pepper.jpeg"
col_img = cv2.imread(img_path)
cv2_imshow(col_img) 
bnw_img = cv2.cvtColor(col_img,cv2.COLOR_BGR2GRAY)
cv2_imshow(bnw_img) 
print(bnw_img.shape)

filter_path = "/content/filter.jpeg"
col_fil = cv2.imread(filter_path)
cv2_imshow(col_fil) 
bnw_fil = cv2.cvtColor(col_fil,cv2.COLOR_BGR2GRAY)
cv2_imshow(bnw_fil) 
print(bnw_fil.shape)

result = cv2.matchTemplate(bnw_img, bnw_fil,	cv2.TM_CCORR_NORMED)

(minVal, maxVal, minLoc, maxLoc) = cv2.minMaxLoc(result)
(startX, startY) = maxLoc
endX = startX + bnw_fil.shape[1]
endY = startY + bnw_fil.shape[0]
cv2.rectangle(bnw_img, (startX, startY), (endX, endY), (255, 0, 0), 3)
cv2_imshow(bnw_img)
```
